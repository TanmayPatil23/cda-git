# json to DataFrame
Converting Json data distributed among 20 different json files into one single pandas 
DataFrame

```python
files = os.listdir()
dataFrames = list()

for file in files:
    if file.endswith('.json'):
        f = open(file)
        data = json.load(f)
        normalised_data = pd.json_normalize(data)
        dataFrame = pd.DataFrame(normalised_data)
        dataFrames.append(dataFrame)
        f.close()
        
data_complete = pd.concat(dataFrames)
chess_data = pd.concat(dataFrames)
```

```python
import itertools
import re
import hashlib

# a shingle in this code is a string with K-words
K = 1

def jaccard_set(s1, s2):
    " takes two sets and returns Jaccard coefficient"
    u = s1.union(s2)
    i = s1.intersection(s2)
    return len(i)/len(u)

def make_a_set_of_tokens(doc):
    """makes a set of K-tokens"""
    tokens = re.sub("[^\w]", " ",  doc).split()

    sh = set()
    for i in range(len(tokens)-K):
        t = tokens[i]
        for x in tokens[i+1:i+K]:
            t += ' ' + x 
        sh.add(t)
    return sh

if __name__ == '__main__':

    documents = chess_data['moves'][0]

    shingles = []
    # handle moves one by one
    # makes a list of sets which are compresized of a list of K words string
    for doc in documents:
        # makes a set of tokens
        # sh = set([' ', ..., ' '])
        sh = make_a_set_of_tokens(doc)
#         print(f"sh = {sh}")

        # hasing
        bucket = map(hash, sh)

#         print(f"bucket = {list(bucket)}")
        
        # shingles : list of sets (sh)
        shingles.append(set(bucket))

    #print("shingles=%s") %(shingles)
    
    combinations = list( itertools.combinations([x for x in range(len(shingles))], 2) )
#     print(f"combinations={combinations}")

    # compare each pair in combinations tuple of shingles
    for c in combinations:
        i1 = c[0]
        i2 = c[1]
        jac = jaccard_set(shingles[i1], shingles[i2])
        print(f"{c}: jaccard={jac}")
```
